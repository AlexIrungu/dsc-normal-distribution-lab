## A quick refresher! 

Here's the formula for the normal distribution density function once more:

$$ \large N(x) = \dfrac{1}{\sigma \sqrt {2\pi }}e^{-\dfrac{(x-\mu)^2}{2\sigma^2}}$$

Here, 
- $\mu$ is the mean
- $\sigma$ is the standard deviation
- $\pi \approx 3.14159 $ 
- $ e \approx 2.71828 $


## First generate a normal distribution containing 5000 values with $\mu=14$ and $\sigma = 2.8$

```python
# Generate a random normal variable with given parameters , n=5000
import numpy as np

# Parameters for the normal distribution
mu = 14  # Mean
sigma = 2.8  # Standard deviation
size = 5000  # Number of values

# Generate the normal distribution
normal_distribution = np.random.normal(mu, sigma, size)
normal_distribution
```

## Calculate a normalized histogram for this distribution in matplotlib, with bin size = 20

Make sure to get the bin positions and counts for each of the obtained bins. You can use [official documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) to view input and output options for `plt.hist()`

```python
# Calculate a histogram for above data distribution
import matplotlib.pyplot as plt
import numpy as np

# Assuming normal_distribution contains your data
mu = 14  # Mean
sigma = 2.8  # Standard deviation
size = 5000  # Number of values

# Generate the normal distribution
normal_distribution = np.random.normal(mu, sigma, size)

# Calculate and plot the normalized histogram
n, bins, patches = plt.hist(normal_distribution, bins=20, density=True, alpha=0.6, color='g')

# Print bin positions and counts
print("Bin edges:", bins)
print("Counts (normalized):", n)

plt.xlabel('Value')
plt.ylabel('Probability density')
plt.title('Normalized Histogram of the Distribution')
plt.show()
```

## Use the formula to calculate the density function with $\mu$, $\sigma$ and bin information obtained before

```python
# Calculate the normal Density function 
import numpy as np

# Assuming mu, sigma, and bins are already defined
mu = 14  # Mean
sigma = 2.8  # Standard deviation
# bins obtained from the histogram

# Calculate the center of each bin
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# Calculate the density function for each bin center
density = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((bin_centers - mu) / sigma) ** 2)
density
```

## Plot the histogram and density function

```python
# Plot histogram along with the density function
import matplotlib.pyplot as plt
import numpy as np



# Plot the normalized histogram
n, bins, patches = plt.hist(normal_distribution, bins=20, density=True, alpha=0.6, color='g', label='Histogram')

# Calculate the center of each bin
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# Calculate the density function for each bin center
density = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((bin_centers - mu) / sigma) ** 2)

# Plot the density function
plt.plot(bin_centers, density, '--', color='r', label='Density Function')

plt.xlabel('Value')
plt.ylabel('Probability density')
plt.title('Normalized Histogram and Density Function')
plt.legend()
plt.show()
```

## Visualize the distribution using seaborn and plot the KDE

```python
# Use seaborn to plot the histogram with KDE
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming normal_distribution contains your data
# Plot the KDE
sns.kdeplot(normal_distribution, bw_adjust=0.5, color='blue', fill=True)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE of the Distribution')
plt.show()
```
